## 1. Minh họa API Gateway định tuyến request đến microservice


```python
service_routes = {
    '/api/v1/products': 'product-service:3001',
    '/api/v1/cart': 'cart-service:3007',
    '/api/v1/orders': 'order-service:3004',
    '/api/v1/auth': 'authentication-service:3008',
}

def route_request(path: str):
    for prefix, service in service_routes.items():
        if path.startswith(prefix):
            return {
                'request_path': path,
                'forward_to': service,
                'message': f'API Gateway chuyển request {path} đến {service}'
            }
    return {'error': 'Không tìm thấy service phù hợp'}

for path in ['/api/v1/products', '/api/v1/cart/items', '/api/v1/orders/123']:
    print(route_request(path))
```

## 2. Minh họa Product Service xử lý danh sách sản phẩm


```python
products = [
    {'id': 'p01', 'name': 'Áo thun basic', 'price': 120000, 'stock': 15},
    {'id': 'p02', 'name': 'Quần jean nam', 'price': 350000, 'stock': 8},
    {'id': 'p03', 'name': 'Áo sơ mi trắng', 'price': 260000, 'stock': 0},
]

def find_all_products():
    return products

def find_available_products():
    return [p for p in products if p['stock'] > 0]

print('Tất cả sản phẩm:')
print(find_all_products())
print('\nSản phẩm còn hàng:')
print(find_available_products())
```

## 3. Minh họa workflow xử lý đơn hàng và gửi thông báo


```python
def handle_order_created(order):
    email_payload = {
        'to': order['customerEmail'],
        'subject': f"Xác nhận đơn hàng #{order['orderId']}",
        'message': f"Xin chào {order['customerName']}, đơn hàng {order['orderId']} trị giá {order['total']} VND đã được ghi nhận."
    }
    return email_payload

order_event = {
    'orderId': 'ORD-001',
    'customerName': 'Nguyễn Văn A',
    'customerEmail': 'customer@example.com',
    'total': 470000,
}

print(handle_order_created(order_event))
```

## 4. Minh họa AI Agent trung tâm điều phối yêu cầu


```python
routing_rules = [
    (['size', 'màu', 'chất liệu', 'còn hàng', 'giá', 'sản phẩm'], 'Product Consulting Agent'),
    (['phối đồ', 'mặc với', 'gợi ý outfit', 'hợp với'], 'Fashion Recommendation Agent'),
    (['giỏ hàng', 'chưa thanh toán', 'bỏ quên'], 'Cart Reminder Agent'),
    (['đơn hàng', 'vận chuyển', 'thanh toán', 'trạng thái'], 'Order Support Agent'),
    (['hành vi', 'xem nhiều', 'mua nhiều', 'thống kê'], 'User Behavior Analysis Agent'),
    (['viết mô tả', 'tạo nội dung', 'email', 'thông báo'], 'Content Generation Agent'),
]

def route_to_agent(message: str):
    message_lower = message.lower()
    for keywords, agent in routing_rules:
        if any(keyword in message_lower for keyword in keywords):
            return agent
    return 'Product Consulting Agent'

samples = [
    'Áo này còn size M không?',
    'Áo sơ mi này phối với quần gì?',
    'Đơn hàng của tôi đang ở đâu?',
    'Viết mô tả ngắn cho sản phẩm áo khoác denim',
]

for s in samples:
    print(f'{s} -> {route_to_agent(s)}')
```
